**Batch learning**

Big Data involves two factors
1. Volume - data has huge volume
2. Speed - data comes with high speed

**Data streams**
* Data arrives in streams
* Data is assumed infinite
* Constant flow of data
* Data arrives rapidly

**Examples of stream soures**
* Sensor data
* Image data
* Internet

**Stream processing**
* Stream must be processed real-time.
* Stream processing must be effcient.



**Online machine learning** is a

* method of machine learning in which data becomes available in a sequential order and is used to update the best predictor for future data at each step 
* as opposed to batch learning techniques which generate the best predictor by learning on the entire training data set at once.

image.png

**Concept drift** is the change in the relationships between input and output data over time.

**Concept drift detection** is the task detection of occurance of concept drift.

We need to detect a concept drift because the model performance can degrade over time.

**Hidden context** is information not present in the data but still affects it.

**Concept drift patterns**
* Sudden drift - a new concept occurs within a short time.
* Gradual drift - a new concept gradually replace an old one over a period of time
* Incremental drift - the old concept incrementally changes to new concept
* Recurring drift - the old concept may recur after some time






image.png

**Addressing Concept Drift**
* Train a new model - train new model on more recent data
* Update the existing model - updating the existing model
* Learn the drift - learn how the underlying distribution is shifting over time, and incorporate this change into the model itself

### Stream data generators

```python
# SEAGenerator can produce a constant stream of data with frequent concept drift
from skmultiflow.data import SEAGenerator
# Create an instance of the generator with 0.28 noise 
stream = SEAGenerator(noise_percentage = 0.28)
# Get a datapoint from the stream
print(stream.next_sample())

# The number of remaining samples is -1
# because the generator concept implies 
# that infinite samples can be generated
# print("Number of samples",stream.n_remaining_samples())
```

### Model evaluation with stream data

**Prequential error** is the accumulated loss over a sequence of predictions with a sliding window of some length (X last predictions).

The prequential error evaluation is the method of evaluating a model based on the prequential error

image.png

```python
%matplotlib notebook

from sklearn.naive_bayes import MultinomialNB
# Import function for evaluating prequential error
from skmultiflow.evaluation import EvaluatePrequential

classifier = MultinomialNB()

# Function to prequential error over 30000 samples showing accuracy, running time, and model size
eval = EvaluatePrequential(show_plot=True,
                           max_samples=30000,
                           metrics=['accuracy', 'running_time', 'model_size'])

# Execute the prequential error plotting function using the MultinomialNB classifier and the SEAGenerator datastream
eval.evaluate(stream=stream, model=classifier, model_names=['NB']);
```

### Concept drift detection

**ADWIN** (adaptive sliding window) is an adaptive sliding window algorithm for detecting change, and keeping updated statistics about a data stream.

The main idea of ADWIN is to keep statistics from a window of variable size while detecting concept drift. Changes are detected by testing whether the difference between the averages of two sub-windows in window $W$ is larger than a given threshold.


image.png

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Import ADWIN
from skmultiflow.drift_detection import ADWIN

# Let datastream be 2000 values of either 1 or 0
drift_stream = np.random.randint(2, size=2000)

# To introduce a concept drift we change the last 1000 values to either 4, 5, 6, or 7
for i in range(999, 2000):
    drift_stream[i] = np.random.randint(4, high=8)

plt.figure(figsize=(20, 10))
plt.plot(drift_stream)

# instantiate ADWIN object 
adwin = ADWIN()

# for each data point in stream 
for i in range(drift_stream.size):
    # add a new point to adwin object
    adwin.add_element(drift_stream[i])
    
    # if adwin detects change, print at what point in the stream 
    # the change was detected
    if adwin.detected_change():
        print('Change detected at index {}'.format(i))
```
